```python
import selenium
import pandas as pd
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(r"C:\Users\vinay\OneDrive\Desktop\FlipRobo\Selenium\chromedriver_win32\chromedriver.exe")
time.sleep(3)
driver.maximize_window()
```

```python
cities = ['Kolkata', 'Chennai', 'Mumbai', 'Cochin', 'Hyderabad', 'Delhi', 'Ahmedabad', 'Patna', 'Ranchi'
          'Lucknow', 'Pune', 'Jaipur', 'Srinagar', 'Bhubaneswar', 'Bengaluru', 'Chandigarh', 'Visakhapatnam', 'New Delhi']
```

```python
make = 'https://www.makemytrip.com/'
```

```python
def combinantorial(lst):
    count = 0
    index = 1
    pairs = []
    for element1 in lst:
        for element2 in lst[index:]:
            pairs.append((element1, element2))
        index += 1

    return pairs
```

```python
zipped = combinantorial(cities)
```

```python
Airline = []
Date_of_journey = [] 
Source = []
Destination = []
Departure_time = []
Arrival_time = []
Duration = []
Total_stops = [] 
Price = []
```

```python
for p in zipped:
    driver.get(make)
    time.sleep(3)
    try:
        driver.find_element_by_xpath("//div[@class='loginModal displayBlock modalLogin dynHeight personal ']").click()
    except:
        pass
    
    #Clicking input city.
    driver.find_element_by_xpath("//input[@class='fsw_inputField lineHeight36 latoBlack font30']").click()
    time.sleep(1)
    
    #Entering source name
    driver.find_element_by_xpath("//input[@aria-controls='react-autowhatever-1']").send_keys(p[0])
    time.sleep(2)
    
    #Selecting first auto recommendation.
    driver.find_element_by_xpath("//li[@id='react-autowhatever-1-section-0-item-0']/div").click()
    time.sleep(1)
    
    try:
        #Entering destination name
        driver.find_element_by_xpath("//input[@aria-controls='react-autowhatever-1']").send_keys(p[1])
        time.sleep(2)
    except:
        driver.find_element_by_xpath("//label[@for='toCity']").click()
        time.sleep(1)
        driver.find_element_by_xpath("//input[@aria-controls='react-autowhatever-1']").send_keys(p[1])
        time.sleep(2)
    
    #Selecting first auto recommendation.
    driver.find_element_by_xpath("//li[@id='react-autowhatever-1-section-0-item-0']/div").click()
    time.sleep(1)
    
    #Clicking today's date
    driver.find_element_by_xpath("//div[@class='DayPicker-Day DayPicker-Day--selected']/div").click()
    
    #Clicking on search button for results
    search_button = driver.find_element_by_xpath("//a[@class='primaryBtn font24 latoBold widgetSearchBtn ']")
    search_button.click()
    
    time.sleep(3)
    
    #Close pop up.
    try:
        driver.find_element_by_xpath("//span[@class='bgProperties icon20 overlayCrossIcon']").click()
    except:
        pass
    
    try:
        #Airline name
        names = driver.find_elements_by_xpath("//span[@class='boldFont blackText airlineName']")
        for i in names:
            Airline.append(i.text)

        #Departure time
        depatures = driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoLeft']/div/p/span")
        for i in depatures:
            Departure_time.append(i.text)

        #Source
        sources = driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoLeft']/div/p[2]")
        for i in sources:
            Source.append(i.text)

        #arrival time
        arrival = driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoRight']/div/p/span")
        for i in arrival:
            Arrival_time.append(i.text)

        #destina
        destina = driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoRight']/div/p[2]")
        for i in destina:
            Destination.append(i.text)

        #Duration
        durations = driver.find_elements_by_xpath("//div[@class='stop-info flexOne']/p")
        for i in durations:
            Duration.append(i.text)

        #Stops
        stops = driver.find_elements_by_xpath("//p[@class='flightsLayoverInfo']")
        for i in stops:
            Total_stops.append(i.text)

        #Prices
        pricess = driver.find_elements_by_xpath("//p[@class='blackText fontSize18 blackFont white-space-no-wrap']")
        for i in pricess:
            Price.append(i.text)


        #Date of departure
        date_list = []
        date_today = driver.find_element_by_xpath("//div[@class='slider-slide slide-visible']/div/a/p").text
        date_list = [date_today]*len(pricess)
        Date_of_journey.append(date_list)
    
    
        try:
            #To get a months data
            for z in range(2, 7):
                driver.find_elements_by_xpath("//div[@class='slider-list']/div/div")[z].click()
                time.sleep(5)

                #Close pop up.
                try:
                    driver.find_element_by_xpath("//span[@class='bgProperties icon20 overlayCrossIcon']").click()
                except:
                    pass

                #Airline name
                names = driver.find_elements_by_xpath("//span[@class='boldFont blackText airlineName']")
                for i in names:
                    Airline.append(i.text)

                #Departure time
                depatures = driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoLeft']/div/p/span")
                for i in depatures:
                    Departure_time.append(i.text)

                #Source
                sources = driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoLeft']/div/p[2]")
                for i in sources:
                    Source.append(i.text)

                #arrival time
                arrival = driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoRight']/div/p/span")
                for i in arrival:
                    Arrival_time.append(i.text)

                #destina
                destina = driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoRight']/div/p[2]")
                for i in destina:
                    Destination.append(i.text)

                #Duration
                durations = driver.find_elements_by_xpath("//div[@class='stop-info flexOne']/p")
                for i in durations:
                    Duration.append(i.text)

                #Stops
                stops = driver.find_elements_by_xpath("//p[@class='flightsLayoverInfo']")
                for i in stops:
                    Total_stops.append(i.text)

                #Prices
                pricess = driver.find_elements_by_xpath("//p[@class='blackText fontSize18 blackFont white-space-no-wrap']")
                for i in pricess:
                    Price.append(i.text)


                #Date of departure
                date_list = []
                date_today = driver.find_element_by_xpath("//div[@class='slider-slide slide-visible']/div/a/p").text
                date_list = [date_today]*len(pricess)
                Date_of_journey.append(date_list)
                time.sleep(2)
        except:
            pass
    except:
        pass
```

```python
flat_list = [item for sublist in Date_of_journey for item in sublist]
```

```python
flight_price = pd.DataFrame({})

flight_price['Airline'] = Airline
flight_price['Date_of_journey'] = flat_list
flight_price['Source'] = Source[:-10]
flight_price['Destination'] = Destination[:-10]
flight_price['Departure_time'] = Departure_time[:-10]
flight_price['Arrival_time'] = Arrival_time[:-10]
flight_price['Duration'] = Duration[:-10]
flight_price['Total_stops'] = Total_stops[:-10]
flight_price['Price'] = Price
```

```python
flight_price.to_excel("Flight_price_scraped.xlsx")
```
